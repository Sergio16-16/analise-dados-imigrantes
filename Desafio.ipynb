```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df = pd.DataFrame(vendas_2022)
```

```python
df.index = ['Loja A', 'Loja B', 'Loja C', 'Loja D']
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
         'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

print (meses)
```

```python
print(df)
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigrao dos quatro maiores pases da Amrica do Sul para o Canad de 1980 a 2013')

axs[0,0].plot(df.loc['Loja A', meses])
axs[0,0].set_title('Loja A')

axs[0,1].plot(df.loc['Loja B', meses])
axs[0,1].set_title('Loja B')

axs[1,0].plot(df.loc['Loja C', meses])
axs[1,0].set_title('Loja C')

axs[1,1].plot(df.loc['Loja D', meses])
axs[1,1].set_title('Loja D')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(1))

for ax in axs.flat:
  ax.set_xlabel('Meses')
  ax.set_ylabel('Nmero de vendas')

ymin = 0
ymax = 500

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```
