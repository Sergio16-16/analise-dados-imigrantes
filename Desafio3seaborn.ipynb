```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Usuário/Downloads/canadian_immegration_data.csv")
```

```python
top_4 = df[df['Country'].isin(['Brazil', 'Argentina', 'Colombia', 'Peru'])].sort_values('Total', ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.barplot(data=top_4, y='Country', x='Total', orient='h', hue =top_4.index, palette='tab10', legend = False)

ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
sns.despine()
ax.set_ylabel('') 
```

```python
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_4, y='Country', x='Total', orient='h', hue=top_4.index, palette=palette, legend=False)
    ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
    ax.set_xlabel('Número de imigrantes', fontsize=14)
    ax.set_ylabel('')
    plt.show()
```

```python
gerar_grafico_paleta('coolwarm')
```

```python
gerar_grafico_paleta('Blues_r')
```

```python
gerar_grafico_paleta('rocket')
```
