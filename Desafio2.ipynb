```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df = pd.DataFrame(vendas_2022)
```

```python
df.index = ['Loja A', 'Loja B', 'Loja C', 'Loja D']
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
         'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

print (meses)
```

```python
print(df)
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas', fontsize = 20, color = 'royalblue')

lojas = ['Loja A', 'Loja B', 'Loja C', 'Loja D']

for ax, loja in zip(axs.ravel(), lojas):
    dados = df.loc[loja, meses]
    
    ax.plot(dados, color='forestgreen')
    
    # pico
    max_valor = dados.max()
    mes_pico = dados.idxmax()
    
    # bolinha
    ax.scatter(mes_pico, max_valor, color='red', s=100, zorder=3)
    ax.annotate(f'{max_valor}',
            (mes_pico, max_valor),
            textcoords="offset points",
            xytext=(0,10),
            ha='center')
    
    ax.set_title(loja)
    ax.set_frame_on(False)
    ax.xaxis.set_tick_params(labelsize=12)
    ax.yaxis.set_tick_params(labelsize=12)


for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(1))

for ax in axs.flat:
  ax.set_xlabel('Meses')
  ax.set_ylabel('Número de vendas')

ymin = 0
ymax = 500


for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)
fig.savefig('grafico_vendas.png', dpi=300, bbox_inches='tight')
plt.show()
```
